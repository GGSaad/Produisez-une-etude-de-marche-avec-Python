```python
# Utiliser la fonction pivot_table pour agréger la population par zone
pop_2017 = pd.pivot_table(pop_2017, index='Zone', values='Valeur', aggfunc='sum')

# Réinitialisez l'index pour avoir une DataFrame régulière
pop_2017 = pop_2017.reset_index()
# Conversion des valeurs de la colonne 'Valeur' en entiers
pop_2017['Valeur'] = pop_2017['Valeur'].astype(int)
# Renommer la colonne 'Valeur' en 'population'
pop_2017 = pop_2017.rename(columns={'Valeur': 'population'})

# Afficher les premières lignes de la DataFrame mise à jour
pop_2017.head()
```

### Dispo_alimentaire2017

```python
# choisir les colonne nécessaires
Dispo_Alimentaire=aliments[['Zone','Élément','Produit','Valeur','Unité']]
Dispo_Alimentaire
```

```python
# Définir les indicateurs pertinents pour l'analyse
pertinent =['Disponibilité intérieure','Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)'
             ,'Disponibilité de protéines en quantité (g/personne/jour)', 'Importations - Quantité',
             'Exportations - Quantité','Production','Nourriture']
dispo_2017= pd.pivot_table(aliments[aliments['Élément'].isin(pertinent)],
                                                      index ='Zone', values= "Valeur", aggfunc= 'sum', columns=['Élément'])
dispo_2017.info() 
```

```python
dispo_2017
```

```python
dispo_2017.isnull().sum()
```

```python
Dispo_Alimentaire['Produit'].unique()
```

```python
Dispo_Poulet=Dispo_Alimentaire.loc[Dispo_Alimentaire['Produit']=='Viande de Volailles'].reset_index()
Dispo_Poulet
```

```python
Dispo_Poulet=Dispo_Poulet.pivot_table(index='Zone',columns='Élément',values='Valeur').reset_index()
```

```python
Dispo_Poulet=Dispo_Poulet[['Zone','Disponibilité alimentaire en quantité (kg/personne/an)','Disponibilité intérieure','Exportations - Quantité','Importations - Quantité','Nourriture','Production']]
Dispo_Poulet
```

```python
Dispo_Poulet=Dispo_Poulet.rename(columns={'Disponibilité alimentaire en quantité (kg/personne/an)':'Dispo Poulet en quantité (kg/personne/an)','Disponibilité intérieure': 'Dispo intérieure poulet', 
                        'Exportations - Quantité':'Exportations Poulet',
                        'Importations - Quantité':'Importations Poulet', 
                        'Production':'Production Poulet','Nourriture':'Consommation Poulet'})
```

```python
Dispo_Poulet=pd.merge(Dispo_Poulet,dispo_2017, on='Zone',how='left' )
Dispo_Poulet
```

```python
#produire des nouveau indicateures rélaives aux poulets à partir des indicateurs totales
Dispo_Poulet['Taux_Import_P%']= round((Dispo_Poulet['Importations Poulet']*100/ Dispo_Poulet['Importations - Quantité']),2)
Dispo_Poulet['Taux_Export_P%']= round((Dispo_Poulet['Exportations Poulet']*100/ Dispo_Poulet['Exportations - Quantité']),2)
Dispo_Poulet['Taux_Conso_P%']= round((Dispo_Poulet['Consommation Poulet']*100/ Dispo_Poulet['Nourriture']),2)
```

```python
Dispo_Poulet
```

```python
# Garder les données des poulets
Dispo_Poulet=Dispo_Poulet[['Zone','Dispo Poulet en quantité (kg/personne/an)','Dispo intérieure poulet','Exportations Poulet','Importations Poulet','Consommation Poulet','Production Poulet','Taux_Import_P%','Taux_Export_P%','Taux_Conso_P%']]
# nombre de pays dans cette dataframe
Dispo_Poulet['Zone'].nunique()
```

### Coût et abordabilité d’une alimentation saine (CoAHD)

```python
# Sélection des colonnes "Zone", "Produit", et "Valeur"
COAHD = COAHD[['Zone', 'Produit', 'Valeur']]

# Filtrage pour ne conserver que les lignes avec la valeur "Cost of animal source foods" dans la colonne "Produit"
COAHD = COAHD[COAHD['Produit'] == 'Cost of a healthy diet (PPP dollar per person per day)']


# Réinitialisation de l'index
COAHD = COAHD.reset_index(drop=True)

# Sélection des colonnes "Zone" et "Valeur"
COAHD = COAHD[['Zone', 'Valeur']]

# Affichage du DataFrame après les opérations de sélection et de filtrage
COAHD.head()
```

```python
#renommage perinent de la colonne Valeur
COAHD=COAHD.rename(columns={'Valeur':'cout_alimentation_saine/p'})
# Converir les valeurs car ils ont été donné par jour
#COAHD["cout_poulet"] =round(population["cout_poulet"]*365,2) #
#COAHD.head()
```

```python
COAHD["cout_alimentation_saine/p"] =round(COAHD["cout_alimentation_saine/p"]*365,2)
COAHD
```
